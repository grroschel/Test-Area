```python
import pandas as pd
```

```python
# Create a DataFrame
pay_dict = {
    "Name": ["Greg", "Dave", "Scott", "Robert"],
    "Salary": [10000, 9000, 8000, 7000],

}
pay_data = pd.DataFrame(pay_dict)
print(f"There are {len(pay_data)} records in our Pay Data Data Frame")
pay_data
```

```python
# Add income tax
tax = pay_data["Salary"] * .1
pay_data["Income Tax"] = tax
pay_data
```

```python
# Add property tax
prop_tax = pay_data["Salary"] * .5
pay_data["Property Tax"] = prop_tax
pay_data
```

```python
# Add net income
net_income = pay_data["Salary"] - (pay_data["Income Tax"] + pay_data["Property Tax"])
pay_data["Net Income"] = net_income
pay_data
```

```python
print(f"The data types are:")
print(f"-------------------")
print(pay_data.dtypes)

# Convert the numeric columns to numeric so the data can be used
pay_data["Income Tax"] = pd.to_numeric(pay_data["Income Tax"]) 
pay_data["Property Tax"] = pd.to_numeric(pay_data["Property Tax"]) 
pay_data["Net Income"] = pd.to_numeric(pay_data["Net Income"]) 

pay_data["Salary"] = pay_data["Salary"].map("${:,.2f}".format)
pay_data["Income Tax"] = pay_data["Income Tax"].map("${:,.2f}".format)
pay_data["Property Tax"] = pay_data["Property Tax"].map("${:,.2f}".format)
pay_data["Net Income"] = pay_data["Net Income"].map("${:,.2f}".format)
pay_data                                        

pay_data = pay_data.set_index("Name")
pay_data

pay_data = pay_data.rename(columns={"Income Tax": "Income Tax (10%)",
                                    "Property Tax" : "Property Tax (50%)"
                                   })
print()
print("The final Pay Data Data Frame:")
print("------------------------------")
pay_data
```
